````python
date = pd.to_datetime("2021-02-25") #Filtro da data

df_states_filter = df_states[df_states["date"] == date].sort_values(by = "vaccinated_second", ascending = False)

top_10_imu = df_states_filter[1:11][["state","vaccinated_second"]] # começa com o index 1 pois o primeiro valor é o total

top_10_imu.reset_index().drop(columns=['index'])
````
